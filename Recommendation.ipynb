```python
food2 = pd.read_csv(data_url, usecols=['title', 'rating', 'calories', 'fat','sodium'])
```

```python
food2["Unique_id"] = food2.index + 1
```

```python
food2.sample()
```

```python
food2.dropna()
```

```python
combine_food_rating = food2.dropna(axis = 0, subset = ['title'])
food_ratingCount = (food2.
     groupby(by = ['title'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['title', 'totalRatingCount']]
    )
food_ratingCount.sample(10)
```

```python
food2 = food2.merge(food_ratingCount, left_on = 'title', right_on = 'title', how = 'left')
```

```python
food2.sample()
```

```python
food2.describe()
```

```python
popularity_threshold = 5
rating_popular_food= food2.query('totalRatingCount >= @popularity_threshold')
rating_popular_food.head()
```

```python
rating_popular_food.shape
```

```python
rating_popular_food.isna().value_counts()
```

```python
food_features_df=rating_popular_food.pivot_table(index='title',columns='Unique_id',values='rating').fillna(0)
food_features_df.head()
```

```python
from scipy.sparse import csr_matrix

food_array_matrix = csr_matrix(food_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(food_array_matrix)
```

```python
query_index = np.random.choice(food_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(food_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
food_features_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(food_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, food_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
